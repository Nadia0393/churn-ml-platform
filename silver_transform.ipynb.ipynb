```python
df_bronze = spark.table("bronze_churn")
```

```python
df_bronze.printSchema()
```

```python
from pyspark.sql.functions import col,when
from pyspark.sql.types import DoubleType, IntegerType

df_silver=df_bronze.withColumn("TotalCharges",when(col("TotalCharges")==" ",None).otherwise(col("TotalCharges")))
```

```python
df_silver = df_silver.withColumn(
    "TotalCharges",
    col("TotalCharges").cast(DoubleType())
)
```

```python
df_silver = df_silver.withColumn(
    "tenure", col("tenure").cast(IntegerType())
)
```

```python
df_silver = df_silver.withColumn(
    "MonthlyCharges", col("MonthlyCharges").cast(DoubleType())
)

df_silver = df_silver.withColumn(
    "SeniorCitizen", col("SeniorCitizen").cast(IntegerType())
)
```

```python
df_silver=df_silver.dropna(subset=["TotalCharges"])
```

```python
print("Silver row count:", df_silver.count())
```

```python
spark.sql("DROP TABLE IF EXISTS silver_churn")
df_silver.write.format('delta').saveAsTable("silver_churn")
```
